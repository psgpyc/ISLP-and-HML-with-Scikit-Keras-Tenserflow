# Receiver Operating Characteristic (ROC) Curve

## 1. Introduction  
The **Receiver Operating Characteristic (ROC) curve** is a graphical representation of a classification model’s performance across different decision thresholds. It helps evaluate the trade-off between **True Positive Rate (Recall)** and **False Positive Rate (FPR)**.

---

## 2. Key Metrics  

### True Positive Rate (TPR) – Recall  

TPR = $\frac{TP}{TP + FN}$

- Measures how well the model captures actual positives.

### False Positive Rate (FPR)  

FPR = $\frac{FP}{FP + TN}$

- Measures how many negatives are misclassified as positives.

---

## 3. How ROC Curve Works  
- **X-axis:** False Positive Rate (FPR)  
- **Y-axis:** True Positive Rate (TPR or Recall)  
- The curve is plotted by adjusting the classification threshold.

### Interpreting the Curve  
- **A perfect model** has a curve that hugs the **top-left corner** (high TPR, low FPR).  
- **A random model** results in a diagonal line (**y = x**).  
- **A poor model** performs below the diagonal, meaning it's worse than random guessing.

---

## 4. Area Under the Curve (AUC-ROC)  
The **AUC (Area Under the Curve)** represents the model’s overall ability to distinguish between classes.  

- **AUC = 1** → Perfect classifier.  
- **AUC = 0.5** → Random classifier.  
- **AUC < 0.5** → Worse than random (model is making opposite predictions).  

---

## 5. Example  
Imagine a model predicting whether a patient has a disease:  
- A **low threshold** might classify more patients as “diseased,” increasing **TPR** but also **FPR**.  
- A **high threshold** might classify fewer patients as “diseased,” decreasing **FPR** but also reducing **TPR**.  

By plotting these values, we get the **ROC curve**, which helps in choosing the optimal threshold.

---

## 6. Python Code Example  
Here’s a Python example to generate an ROC curve using `sklearn`:

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import roc_curve, auc
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.datasets import make_classification

# Generate synthetic dataset
X, y = make_classification(n_samples=1000, n_features=20, random_state=42)

# Split into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Train model
model = RandomForestClassifier()
model.fit(X_train, y_train)

# Predict probabilities
y_scores = model.predict_proba(X_test)[:, 1]

# Compute ROC curve
fpr, tpr, _ = roc_curve(y_test, y_scores)
roc_auc = auc(fpr, tpr)

# Plot ROC Curve
plt.figure(figsize=(8,6))
plt.plot(fpr, tpr, color='blue', lw=2, label=f'ROC curve (AUC = {roc_auc:.2f})')
plt.plot([0, 1], [0, 1], color='grey', linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve')
plt.legend(loc="lower right")
plt.show()

````python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import roc_curve, auc
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.datasets import make_classification

# Generate synthetic dataset
X, y = make_classification(n_samples=1000, n_features=20, random_state=42)

# Split into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Train model
model = RandomForestClassifier()
model.fit(X_train, y_train)

# Predict probabilities
y_scores = model.predict_proba(X_test)[:, 1]

# Compute ROC curve
fpr, tpr, _ = roc_curve(y_test, y_scores)
roc_auc = auc(fpr, tpr)

# Plot ROC Curve
plt.figure(figsize=(8,6))
plt.plot(fpr, tpr, color='blue', lw=2, label=f'ROC curve (AUC = {roc_auc:.2f})')
plt.plot([0, 1], [0, 1], color='grey', linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve')
plt.legend(loc="lower right")
plt.show()
````

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
```

```python
iris = load_iris(as_frame=True)
```

```python
X = iris.data[['petal width (cm)']].values
```

```python
y = iris.target_names[iris.target] == "virginica"
```

```python
z = iris.target
```

```python
z
```

```python
Xtrain, Xtest, ytrain, ytest = train_test_split(X, y, random_state=42)
```

```python
log_res = LogisticRegression(random_state=42)
```

```python
log_res.fit(Xtrain, ytrain)
```

```python
X_new = np.linspace(0,3, 1000).reshape(-1, 1)
```

```python
y_proba = log_res.predict_proba(X_new)
```

```python
y_proba
```

```python
X_new[y_proba[:,1] >= 0.5][0,0]
```

```python
plt.plot(X_new, y_proba[:,1], label="True label")
plt.plot(X_new, y_proba[:,0], label="False label")

plt.legend()
```

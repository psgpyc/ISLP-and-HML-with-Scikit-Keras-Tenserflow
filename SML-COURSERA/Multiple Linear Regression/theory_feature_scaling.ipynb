## Feature Scalling



If different features takes on differe ranges of values, it will cause gradient descent to converge slowly. Therefore, scaling each feature to similar range will make the convergance faster.

## How do we scale feature?


One way of doing so is:

if we have two different features:

$ 300 \leq x_1 \leq 200$ and $ 0 <= x_2 <= 5$

Then, we have divided them with the maximum value:

$ x_{1, scaled} = \frac{x_1}{2000}$   

therefore, $ 0.15 \leq x_{1, scaked} \leq 1$

Similarly for feature $x_2$:

$0 \leq x_{2, scaled} \leq 1$

### Mean Normalization

$ x_n = \frac{x_n - \mu}{max(x_n)-min(x_n)}$





### Z-Score Normalization

$ x_n = \frac{x_n - \mu}{\sigma}$





Rule of thumb, aim for about $-1 \leq x_j \leq 1$ for each feature $x_j$

**Acceptable Range**: no need for scaling

$-3 \leq x_j \leq 3$

$-0.3 \leq x_j \leq 0.3$

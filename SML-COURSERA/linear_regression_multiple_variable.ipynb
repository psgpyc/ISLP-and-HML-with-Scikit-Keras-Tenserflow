##  Multiple Feature 

**Notation**

$ x_j = j^{th} \ feature$

$ n = number \ of \ features$

$ \vec{x}^{(i)} = i^{(th)} \  training \ example $

$\vec x_j^{(i)} = value \ of \ feature \ j \ in \ i^{(th)} \ training \ example$

## Updated Model

$f_{(w,b)}(x) =  b + w_1 x_1 + w_2 x_2 + ........ + w_n x_n$

Lets define, a $vector \ w$ for all the parameters

$\vec w = [w_1, w_2, w_3, .... w_n]$

$\vec x = [x_1, x_2, x_3,......., x_n]$

## Algebrically, 

$f_{(w,b)}(x)= \vec w \cdot \vec x + b$

## Vectorisation 

```python
import numpy as np

w = np.array([1,2.5, -3.3])
b = 4
x = np.array([10,20,30])

np.dot(w,x) + b
```

.

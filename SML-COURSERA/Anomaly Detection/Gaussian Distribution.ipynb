# Normal Distribution

## What is Normal Distribution?

The **normal distribution**, also known as the **Gaussian distribution**, is a continuous probability distribution that is symmetric about its mean. It is one of the most commonly used distributions in statistics and is characterized by its bell-shaped curve.

The probability density function (PDF) of the normal distribution is given by:

$$
f(x) = \frac{1}{\sqrt{2\pi\sigma^2}} e^{-\frac{(x - \mu)^2}{2\sigma^2}}
$$

Where:
- \( $\mu$ \): Mean (center of the distribution)
- \( $\sigma$ \): Standard deviation (controls the spread of the distribution)
- \( $\sigma^2$ \): Variance

---

## Properties of Normal Distribution

1. **Symmetry**: The distribution is perfectly symmetrical about the mean (\( $\mu$ \)).
2. **Mean, Median, and Mode**: In a normal distribution, these three measures are equal.
3. **Bell-Shaped Curve**: The shape of the curve is determined by \( $\mu$ \) and \( $\sigma$ \).
   - The mean determines the location of the peak.
   - The standard deviation determines the width of the curve.
4. **68-95-99.7 Rule** (Empirical Rule):
   - 68% of the data lies within 1 standard deviation of the mean (\( $\mu \pm \sigma$ \)).
   - 95% of the data lies within 2 standard deviations of the mean (\( $\mu \pm 2\sigma$ \)).
   - 99.7% of the data lies within 3 standard deviations of the mean (\( $\mu \pm 3\sigma$ \)).

---

## Applications of Normal Distribution

- **Natural Phenomena**: Heights, weights, test scores, etc., often follow a normal distribution.
- **Statistical Methods**: Many statistical tests assume normality.
- **Central Limit Theorem**: The sampling distribution of the sample mean approaches a normal distribution as the sample size increases, regardless of the population's distribution.

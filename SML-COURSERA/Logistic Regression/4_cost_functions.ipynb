## How do we now choose the parameters $\vec w$ and $b$ ?

We have **Squared error cost function** for the linear regression.

$J(\vec w,b) = \frac{1}{m} \sum_{i=1}^{m}\frac{1}{2}(f_{(\vec w, b)} {(\vec x^{(i)} - y^(i))^2})$


We know that, the $J(\vec w,b)$ for $ f_{\vec w,b} (\vec x) = \vec w \cdot \vec x + b$ is a convex cost funciton and only has one global minima. 

Therefore, *Gradient Descent* is garunteed to reach global minimum.

### Why the same Squared Error Cost Function wont work for logistic Regression?

But,  the $J(\vec w,b)$ for $ f_{\vec w,b} (\vec x) = \frac{1}{1 + e^{-(\vec w \cdot \vec x + b)} }$
is a non convex cost funciton and has a lot of local minima.


## We will use a diffent function which has a the cost function convex again.


We will use a loss function, which is a function of prediction and the true label.

$L(f_{\vec w,b}(\vec x^{(i)}), y^{(i)})$ = {

${ -log(f_{\vec w,b}(\vec x^{(i)}) \ if \  y^{(i)} \ = \ 1}$ 

$ { -log(1 - f_{\vec w,b}(\vec x^{(i)}) \ if \  y^{(i)} \ = \ 0} $

}

## The cost function for the logistic regression is:

$ J(\vec w, b) = \frac{1}{m} \sum_{i=1}^m L(f_{\vec w,b}(\vec x^{(i)}), y^{(i)})$

$\frac{True Positive}{ \# Predicted Positives}$ = $\frac{\text{True Positive}}{\text{True Positive} + \text{False Positive}}$

## Recall

#### Of all the predictions we have made for y = 1, what % are Truely detected as 1?

$\frac{\text{True Positive}}{\text{\# Actual Positive}}$ = $\frac{\text{True Positive}}{\text{True Positive + False Negative}}$

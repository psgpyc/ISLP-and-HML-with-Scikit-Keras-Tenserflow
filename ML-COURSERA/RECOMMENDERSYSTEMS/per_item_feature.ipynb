### A Simple Linear Regression Approach.

If we have $W$ and $b$ for each of the user. Then we can make predictions about the ratings for the movies which they have not rated. 

Conside the model,

$ W^{(j)} \cdot X^{(i)} + b^{(j)}$

Here, $j$ superscript for parameters represents a user. 

If we have parameters, $W$ and $b$ for user 1, then we can predict the rating of movie 3.

$W^{(1)} = [5, 0]$  
$b^{(1)} = 0 $  
$x^{(3)} = [0.99, 0]$  

Then, 
$ W^{(1)} \cdot x^{(3)} + b^{(1)}$ = $4.95$


Which seems quiet reasonable, as the user has given 5 rating to movie 1, which is of same genre.


### Cost Function

Here, 

$r(i,j) =1, \text{if \ user \ j \ has \ rated \ the \ movie \ i}$  
$y^{(i,j)} = \text{rating \ given \ by \ user \ j \ to \ movie \ i}$  
$w^{(j)}, b^{(j)} = \text{parameters for user} \ j$    
$x^{(i)} = \text{feature vector for movie i}$  
$m^{(j)} = \text{no. of movies rated by the user j}$    

The cost function will be, 

$J(w^{(j)}, b^{(j)})$  = $\frac{1}{2} \sum_{i:r(i,j)=1} (w^{(j)} \cdot x^{(i)} + b^{(j)} - y^{(i,j)})^2$ + $\frac{\lambda}{2}\sum_{k=1}^{n} (w_k^{(j)})^2$

Here, $n = \text{no. of features}$  
$j = \text{user}$

For all the movies the user has rated, make predictions using the values of $W$ and $b$ and subtract it with the actual rating the user has given. 

Now for all the users,

$\frac{1}{2} \sum_{j=1}^{n_u}$$\frac{1}{2} \sum_{i:r(i,j)=1} (w^{(j)} \cdot x^{(i)} + b^{(j)} - y^{(i,j)})^2$ + $\frac{\lambda}{2}\sum_{j=1}^{n_u}\sum_{k=1}^{n} (w_k^{(j)})^2$

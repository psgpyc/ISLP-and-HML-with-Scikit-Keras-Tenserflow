## The Threshold

How are we going to pick a threshold which defines the classes 1(above the threshold) and 0(below the threshold). 

Mathematically, 

If $f_{(\vec w, b)} (\vec x) \geq 0.5$, then $\hat y = 1$ 

If $f_{(\vec w, b)} (\vec x) < 0.5$, then $\hat y = 0$

## Intuition

**When is $f_{(\vec w, b)} {(\vec x)} \geq 0.5$ ?**

Remeber the sigmoid curve?



When $g(z) > 0$

**When is $g(z) > 0.5 $ ?**

when, $z \geq 0$


**When is $ z \geq 0 $ ?**

when, $\vec w \cdot \vec x + b \geq 0$

## Summing it up:

The model will predict $1$ when, $\vec w \cdot \vec x + b \geq 0$

and, 

The model will predict $0$ when, $\vec w \cdot \vec x + b < 0$

Consider the below dataset,


```python
X = np.array([[0.5, 1.5], [1,1], [1.5, 0.5], [3, 0.5], [2, 2], [1, 2.5]])
y = np.array([0, 0, 0, 1, 1, 1]).reshape(-1,1) 
```

```python
fig, ax = plt.subplots(figsize=(6,5))
ax.scatter(x0[:3], x1[:3], marker='o', label="y=1")
ax.scatter(x0[3::], x1[3:], marker='x', label="y=0")

ax.set_xlabel("$x_0$")
ax.set_ylabel("$x_1$")

ax.legend()
```

Suppose that, we will like to train a logistic regression model on this data which has the form, 

$f(x) = g(w_0x_0 + w_1x_1 + b)$

where $g(z) = \frac{1}{1+e^{-z}}$, which is a sigmoid function

Cossider that, we trained our model and got the parameters $b = -3, w_0=1, w_1=1$, that is

$f(x) = g(x_0+x_1-3)$

this is our logistic regression model.

Our decision boundary is given by, 

$ x_0 + x_1 -3 =0$

which is equivalent to, 

$x_1 = 3 - x_0$

```python
# choosing the value for x0

x_0 = np.arange(6)
x_1 = 3 - x0
```

```python
fig, ax = plt.subplots(1,1, figsize=(6,5))
ax.plot(x_0, x_1)
ax.scatter(x0[:3], x1[:3], marker='o', label="y=1")
ax.scatter(x0[3::], x1[3:], marker='x', label="y=0")

ax.set_xlabel("$x_0$")
ax.set_ylabel("$x_1$")

ax.legend()
```

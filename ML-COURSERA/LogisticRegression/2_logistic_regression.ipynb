### Sigmoid Function / Logistic Function

The sigmoid function looks like below:

 

 The sigmoid function outputs a value between 0 and 1. The formula is:
 

 $g(z) = \frac{1}{1+ e^{-z}}           ,0 < g(z) < 1$

Here, $e$, *eulers number* is a mathematical constant whose value is approximately equal to $2.718$

In the equation we can see that, 

If z is a large +ve number than the output from sigmoid function will be very close to 1.
If z is a large -ve number than the output from sigmoid function will be very close to 0.

And, if $z = 0$ then, 

$ g(0) = \frac{1}{1+1} = \frac{1}{2} = 0.5$

Therefore, the curve passes the vertical axis at 0.5

## Logistic Regression Formula

Consider a straight line function, 

$ f_{(\vec w, b)} = \vec w \cdot \vec x + b$

Let, 



$ z = f_{\vec w, b}  =  \vec w \cdot \vec x + b $

Substuting $z$ into sigmoid function, we will get the logistic regression model

$ f_{(\vec w, b)} = g(\vec w \cdot \vec x + b)$ $= \frac{1}{1+ e^{-(\vec w \cdot \vec x + b)}} $

## Interpretation of logistic regression output:

$ f_{(\vec w, b)} = \frac{1}{1+ e^{-(\vec w \cdot \vec x + b)}}$

Consider, 
$x$ is "tumor size"

$y$ is 0 (not malignant)
       1 (malignant)


Then, 

$f_{(\vec w, b )} {(\vec x)} = 0.7$

means, $70%$ change that true label $y$ is 1.

## Probabilistic defination

$ f_{\vec w, b} {(\vec x)} = P({y = 1} | \vec x; \vec w, b) $

Read as: 

**Probability that $y$ is $1$,**

**given input $\vec x$, parameters $\vec w, b$**

## Implementing Sigmoid Function

**Using Numpy `exp()` function**

```python
import numpy as np
import matplotlib.pyplot as plt

arr = np.array([1,2,3,4,5])

exp_arr = np.exp(arr)
exp_arr
```

```python
def calculate_sigmoid(z):
    g = 1/(1+np.exp(-z))

    return g
```

```python
z_temp = np.arange(-10, 11)
```

```python
y = calculate_sigmoid(z_temp)
```

```python
y
```

```python
fig, ax = plt.subplots(figsize=(5,4))
ax.plot(z_temp, y, c="b")
ax.set_title("Sigmoid function")
ax.set_ylabel('sigmoid(z)')
ax.set_xlabel('z')
```

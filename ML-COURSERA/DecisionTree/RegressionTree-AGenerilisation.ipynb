## Regression Tree

Since it is a regression task, instead of looking into entropy reduction we will now focus onto reducing the variance.

Choosing a split will be similar, but instead we will use variance as a base.




1. Find the variance at the root node
2. Split the examples based on the features
3. Calculate
   1. Variance after split
   2. fraction of examples in each branch
   3. Calculate reduciton of variance ($\text{variance at root} - \text{weighted variane}$)
4. Choose the feature with the largest reduction in variance.
5. Keep on splitting until you meet the criteria.

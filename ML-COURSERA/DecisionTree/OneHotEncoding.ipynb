## One Hot Encoding

If a categorical feature can take on *k* values, create *k* binray features (0 or 1 valued)

We can use one hot encoding for neural networks or any other machine learning algortihms.

## Continuous Valued Features

How are we going to split on something that has continuous values



Steps:

1. Choose different values to split on
2. Carry out calculation and find information gain
3. Chose the value with maximum information gain

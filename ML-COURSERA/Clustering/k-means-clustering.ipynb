## What is Clustering?


The algorithm looks into unlabeled datasets and tries to find clusters within.


Examples:  

- Grouping similar news
- Market Segmentation
- Grouping DNA: DNA Analysis and Segmentation

# k-means Clustering

The very first thing K-means does is find the cluster centroids. And, then it goes to every data points and finds out to which cluster centroid is the point closest to. 

Then it goes into a cluster and finds the average of all the data points in that cluster. Then , move that particular centroid to that avaerage.

It repeats the for as many clusters it has.

## Alogorithm

Randomly initialize $K$ cluster centroids $ \mu_1, \mu_2, \mu_3,.......\mu_k$

where, $\mu_1, \mu_2, \mu_3,.......\mu_k$ are vectors of same dimention as our training examples

i.e. if we have two features ($ n=2 $) these $\mu$ will also be a vector with 2 elements in them.

Repeat {

**Assign points to cluster centroids** 

for $i = 1$ to $ m $  
&nbsp;&nbsp;&nbsp;$ c^{(i)}$ := index(from 1 to $k$) of cluster centroid closest to $x^{(i)}$ 


For a training example $x^{(i)}$, if it is close to $\mu_1$ then $x^{(1)}c^{(1)}$ = $\mu_1$

**Move cluster centroids**

for $k=1$ to $K$
&nbsp;&nbsp;&nbsp; $\mu_k$ := average of points assigned to cluster $k$

}

## Optimisation Algorithm: Cost Function


We have, 

$ c^{(i)}$ = index of cluster($1,2,3......k$) to which example $x^{(i)}$ is currently assigned.  
$\mu_k$ = cluster centroid $k$  
$\mu_c^{(i)}$ = cluster centroid of cluster to which example $x^{(i)}$ has been assigned


Therefore, the **cost function** is,

$$ J(c^{(i)},...., c^{(m)}, \mu_1,....,\mu_k ) = \frac{1}{m}\sum_{i=1}^m || x^{(i)}- \mu_c^{(i)} || $$

In the above chart, for each cluster, the average of the distance between the point and centroid is the value of the cost function

The cost function is also called a **Distortion Function**

# Initialising the K-means

### Random Initialisation

Choose $K$ < $m$  

Randomly pick $k$ training examples.


Set $\mu_1, \mu_2,.....,\mu_k$ equal to these $k$ examples.

**Algorithm using Random Initialisation**

For $i=1$ to $100$ {  
&nbsp;&nbsp; Randomly initialise K-means.  
&nbsp;&nbsp; Run K-means. Get $c^{(1)},...,c^{(m)}, \mu_1,.....,\mu_k$  
&nbsp;&nbsp; Compute cost function (distortion) $J(c^{(i)},...., c^{(m)}, \mu_1,....,\mu_k )$  
}  
Pick set of clusters that gave lowest cost $J$

## Choosing number of clusters

what is the right value of $k$

**Elbow Method**

Start with the Single cluster and calculate the cost function.

Increase the cluser size until it flattens.

Choose the value of k from where the curve started to flatten.

**It is important to note that, choose higher value of k wont work, as higher the cluster, smaller will be the distance between the points in the cluster and its centroid**

### Choosing the value of k

Often, we want to get clusters for some later (downstream) purpose. 

Evaluate K-means based on how well it performs on that later purpose.

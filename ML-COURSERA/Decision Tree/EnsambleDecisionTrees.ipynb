## Sample with replacement & Random Forest Algorithm

Given a training set of size $m$

For $b = 1$ to $B$:  
- Use sampling with replacement to create a new training set of size $m$    
- Train a decision tree on the new dataset

This particular implementatoin of tree ensamble is called **Bagged Decision Tree**.   
We will now look into a better implementation of this algorithm to generate Random Forest Algorithm.


For a length of $B$, We might encounted same features at the root node(or near the root node), we can also randomise this to only select from subset of features.

**Randomising the feature choice**

At each node, when choosing a feature to use to split,  
if $n$ features are available,   
pick a random subset of $k < n$ features and allow the algorithm to only choose from that subset of features.

For a problem with lare no of feature $n$,

the typical implemenattion of $k$ is, 

$$ k = \sqrt n$$


This is Random Forest Algorithm.

## When to use Trees vs Neural Networks

1. **Decision Trees and Tree ensembles**
   - Works well on tabular(structured) data
   - Not recommended for unstuructured data (images, audio, text)
   - Very fast to train
   - Small decision tree many be human interpretable
2. **Neural Network**
   - Works well on all types of data, including tabular (structured) and unstructured data.
   - May be slower than a decision tree
   - works with transfer learning
   - When building a system of multiple models working together, it might be easier to string together multiple neurak networks.

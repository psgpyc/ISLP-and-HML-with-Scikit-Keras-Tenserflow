# Entropy as a meaure of impurity

Consider, 

$p_1 = \text{fraction of examples that are cats}$

So, in a set of 6 animals, if we 3 cats, then $p_1 = \frac{3}{6} = \frac{1}{2}$


The Entropy is given by a function $H(p_{1})$ and has the graph as below:




Now, in our case, the entropy for $\frac{1}{2}$ is $1.0$


So, if we have all the cats, then $H(1)$ is equal to $0.0$

## Lets look at the equation

$H(p_1) = -p_1 log_2(p_1) - p_0 log_2(p_0)$ = $-p_1 log_2(p_1) - (1-p_1)log_2(1-p_1)$


Note: $0 log(0) = 0$

```python
def normalize(X):
    """
    Perform Z-normalisation (Z-score standardisation) on the given dataset.
    
    Z-normalisation scales the data such that it has a mean of 0 and a standard deviation of 1. 
    This method is useful for making features comparable, especially when they are measured on different scales.
    
    Parameters:
    ----------
    X : numpy.ndarray
        A NumPy array containing the data to be normalised. Each column is treated as a feature.

    Returns:
    -------
    numpy.ndarray
        A NumPy array with the same shape as `X`, where each feature is normalised using Z-score standardisation.
    """
    # Compute the mean of each feature (column-wise mean)
    mu = np.mean(X, axis=0)
    
    # Compute the standard deviation of each feature (column-wise standard deviation)
    sigma = np.std(X, axis=0)
    
    # Apply Z-score normalisation: (X - mean) / standard deviation
    znorm = (X - mu) / sigma
    
    # Return the normalised dataset
    return znorm
```

```python
import numpy as np

# Definition:
# The sigmoid function is a mathematical function that maps any real number 
# into a value between 0 and 1. It is commonly used in machine learning, 
# particularly in logistic regression and neural networks, as an activation function.

def sigmoid(a_in_sigmoid):
    """
    Computes the sigmoid of the input.

    Parameters:
        a_in_sigmoid (float or np.array): The input value(s) for which to calculate the sigmoid function.

    Returns:
        float or np.array: The sigmoid value(s) of the input, ranging from 0 to 1.
    """
    # np.exp(-a_in_sigmoid) computes the exponential of the negated input.
    # 1/(1+np.exp(-a_in_sigmoid)) applies the sigmoid formula to the input.
    return 1 / (1 + np.exp(-a_in_sigmoid))
```

```python
import numpy as np

def my_dense(a_in, W, b):
    """
    Computes dense layer with a numerically stable sigmoid activation.
    Args:
      a_in (ndarray (n, )) : Data, 1 example 
      W    (ndarray (n, j)): Weight matrix, n features per unit, j units
      b    (ndarray (j, )) : Bias vector, j units  
    Returns:
      a_out (ndarray (j, )): Output after applying the activation function
    """
    # Compute the weighted sum
    z = np.dot(a_in, W) + b
    
    # Numerically stable sigmoid activation
    a_out = 1 / (1 + np.exp(-z)),
        
    
    return a_out
```

```python
def sequential(X, W1, b1, W2, b2):
    a1 = my_dense(X, W1, b1)
    a2 = my_dense(a1, W2, b2)
    return a2[0]
```

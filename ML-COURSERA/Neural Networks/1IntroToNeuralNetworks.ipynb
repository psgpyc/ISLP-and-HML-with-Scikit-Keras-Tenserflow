## How Neural Networks Work?

In the below example,

We have four features:
1. Price
2. Shipping Cost
3. Marketing Budget
4. Material Used

We are trying to make a demand prediction system using these features, so that we can stock up the product based on their demand. 

Whether a t-shirt becomes a top seller might depends upon few feactors:
1. Affordability
2. Awarness
3. Perceived Quality


Each of the neurones in the below hidden layer(***one or collection of neurones***), will use the input features to predict whether or not the people think t-shirt is affordable and high quality. Also, the middle neurone in the deep layer will use the input feature to determine how aware the customer segment are of the product. 

The output of these there hidden neurones(***activations***) will then be use to predict the probability whether or not the t-shirt will be a top seller




The above neural network has a vector (lets say $ \vec  x$) - a set of input features. 

The input feature is passed to a hidden layer, which outputs another activation vector (lets say $\vec a$)

And, then the activation vector is an input to another layer which then generates an output.


**It is important to note that every input feature is passed on to each hidden layer. The hidden layer than learns to only use relevent feature. For example, the neuron which generates ***affordability activation*** learns to only use input features- price, shipping cost and material and ignore marketing**

## A general Notation to reference units, layers, inputs and the outputs

Let, $l = \text{layer}$  
$j = \text{unit of a layer}$

Then the activation value for a unit can be written as:

$a_j^l = g(\vec w^{[l]}_j \cdot a^{[l-1]} + b_j^{[l]})$

$g() \text{here is a activation function, currently is a sigmoid function}$

**So, for the second layer, unit 3**

$ a^{[2]}_3 = g(W^{[2]}_3 \cdot a^{[2]} + b^{[2]}_3)$

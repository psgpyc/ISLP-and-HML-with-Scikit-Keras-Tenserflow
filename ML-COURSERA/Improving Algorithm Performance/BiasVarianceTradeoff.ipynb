## Bias Variance Tradeoff

1. High Bias: $J_{train}$ is high and $J_{cv}$ is high aswell
2. High variance: $J_{train}$ is low and $J_{cv}$ is high.

3. Just Right:$J_{train}$ is low and $J_{cv}$ is low.

## How regularisation effects bias and varaince tradeoff

### Choosing the value of $\lambda$

Try different values of $\lambda$ and choose the one that gives less cross validation error.

As the $\lambda$ increases, the training cost increases.

## Bias Variance in Neural Networks

#### Large Neural Networks are low bias machines. Obviously!

So, if your model has high error in the train set. Then, we can conclude that it has **high bias.**

One way to resolve this is, to make the NN larger until the test error is reduced to acceptable degree.

---
 
After this, lets check how it does on cross-validation set. Does it permorms well or not?

If it does, that awesome.

If it doesnot, then we have **high variance problem**. And, we know one way to fix high variance proble is to ***get more data***. And, go back and re-train the model.

Limitations: 
1. Expensive to compute larger NN
2. More data, not always available

### It never hurts to choose a larger NN so long we regularise it properly.
